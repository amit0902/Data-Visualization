#***Histogram***

```
%matplotlib inline
plt.hist(df.category_id)
```

```
plt.hist(df.category_id,bins=5,histtype='stepfilled',orientation='vertical',color='r')
plt.xlabel('Category Id')
plt.ylabel('Frequency of Category')
plt.suptitle('Histogram',size=15,y=1.02)
plt.title('For CategoryID of Youtube Video')
plt.savefig('histogram_youtube_categoryid.pdf',Transparent='True')
```

# ***SEABORNE***

```
sns.get_dataset_names()
```

```
e=sns.load_dataset('titanic')
e.head()
```

```
e.shape
```

## ***Strip Plot***

```
sns.stripplot(y='age',data=e,jitter=True)
plt.ylabel('Age in Years')
```

```
sns.stripplot(x='sex',y='age',data=e,jitter=True)
plt.xlabel('Gender')
plt.ylabel('Age in Years')
plt.title('Strip Plot for people Survived in Years')
```

When Jitter= False, We can find range

```
sns.stripplot(x='sex',y='age',data=e,jitter=False)
plt.xlabel('Gender')
plt.ylabel('Age in Years')
plt.title('Strip Plot for people Survived in Years')
```

## ***Swarm Plot***

```
sns.swarmplot(x='class',y='age',data=e)
plt.title('Swarm Plot for Class of people for different age')
```

***Using Swarm Plot for particular category***

```
sns.swarmplot(x='class',y='age',data=e,hue='sex')
```

```
sns.catplot(x='class',y='age',data=e,hue='sex')
```

# ***Box Plot***

```
sns.boxplot(x='who',y='age',data=e)
```

```
sns.violinplot(x='who',y='age',data=e)
```

```
sns.boxplot(x='who',y='age',data=e)
sns.violinplot(x='who',y='age',data=e,color='lightgray')
```

# ***FOR BIVARIATE DATA VISUALIZATION***

## ***JOINT PLOT***

```
sns.jointplot(x='fare',y='age',data=e)
```

### ***DENSITY PLOT***

```
sns.jointplot(x='fare',y='age',data=e,kind="kde")
```

```
sns.jointplot(x='fare',y='age',data=e,kind="hist",color='g')
```

```
sns.jointplot(x='fare',y='age',data=e,kind="hex",color='b')
```

```
sns.jointplot(x='fare',y='age',data=e,kind="reg")
```

```
sns.jointplot(x='fare',y='age',data=e,kind="resid",color='r')
```
